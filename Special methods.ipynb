# **Special Methods in Python / Dunder methods/ Magic Method**

The __init__ method serves as the object constructor in Python, automatically running whenever a new instance of a class is created to initialize its initial attributes and state.

```python
#constructor example 1
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age= age

s1 = Student("Matthew", 30)
print(s1.name)
print(s1.age)
```


### **___str__ method_**

The __str__ method defines the informal, user-friendly string representation of an object, which is automatically triggered when you pass the object to the print() function or the str() cast

```python
#string representation methods - informal
class Student:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return f"student: {self.name}"
s = Student("Matthew")
print(s)
```

### **__repr__ method**

```python
#string representation methods- formal
class Student:
    def __init__(self, name):
        self.name = name
    def __repr__(self):
        return f"Student('{self.name}')"
s = Student("Angelica")
print(repr(s))
```

### **__len__() method**

The __len__ method allows a custom object to mimic a built-in collection by returning its size or element count whenever the global len() function is called on it.

```python
#length method __len__()
class Playlist:
    def __init__ (self, songs):
        self.songs = songs
    def __len__ (self):
        return len(self.songs)
p= Playlist(["Song1", "Song2", "Song3"])
print(len(p))
```

### **__eq__ metho**d

The __eq__ method defines custom equality logic for the == operator, allowing you to compare instances based on their actual internal data rather than their default memory addresses.

```python
#equality comparison __eq__()
class Student:
    def __init__ (self, student_id):
        self.student_id = student_id
    def __eq__(self, other):
        return self.student_id == other.student_id
s1 = Student(101)
s2 = Student(101)
print(s1== s2) 
```


### **__add__ () method**

The __add__ method overloads the addition operator, allowing you to define exactly how two custom objects should interact when the + symbol is used between them.

```python
#operator overloading - addition __add__()
class Vector :
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__ (self, other):
        return Vector(
            self.x + other.x,
            self.y + other.y
        )
    def __str__ (self):
        return f"({self.x}, {self.y})"
v1 = Vector (2,4)
v2 = Vector (4,7)
print(v1 + v2)
```



### **__sub__() method**

The __sub__ method handles the subtraction operator, enabling custom behavior for subtracting one instance from another using the - symbol.

```python
# subtraction __sub__()
class Vector:
    def __init__(self, x):
        self.x= x
    def __sub__ (self, other):
        return Vector(self.x - other.x)
    def __str__ (self):
        return str(self.x)
v1 = Vector(10)
v2 = Vector(6)
print(v1 - v2)
```


### **__mul__() method**

The __mul__ method overloads the multiplication operator, allowing your custom objects to be multiplied by numbers or other instances using the * symbol.

```python
#multiplication __mul__()
class Number:
    def __init__(self, value):
        self.value = value
    def __mul__(self, other):
        return Number (self.value * other.value)
    def __str__(self):
        return str(self.value)
n1= Number (9)
n2= Number (4)
print(n1 * n2)
```




### **__getitem__() method**

The __getitem__ method enables bracket-based indexing or key lookups on an object, allowing you to retrieve data using syntax like obj[key] just like a native list or dictionary.

```python
#item access __getitem__()
class Grades:
    def __init__(self):
        self.data = [85, 90, 78]
    def __getitem__ (self, index):
        return self.data[index]
g= Grades()
print (g[0])
print (g[1])
```



### **__setitem__() method**

The __setitem__ method pairs with item retrieval to allow assignment, enabling you to modify or insert values at a specific index or key using the syntax obj[key] = value.

```python
#item assignment __setitem__()
class Grades:
    def __init__(self):
        self.data = [85, 90,75]
    def __setitem__(self, index, value):
        self.data[index] = value
g = Grades()
g[1] = 100
print(g.data)
```




### **__contains__() method**

This method is used to check if a parameter exists in a class by givng an output of true or false.

```python
#membership testing __contains__()
class Course:
    def __init__(self):
        self.students = ["Ella", "Angela", 'Simon']
    def __contains__ (self, student):
        return student in self.students
c = Course ()
print("Simon" in c)
print("Aisha" in c)
```




### **__call__() method**

The __call__ method allows an instance of a class to be executed exactly like a regular function, enabling syntax like obj() while allowing the underlying object to persist state between executions.

```python
# callable objects __call__ ()
class Multiplier:
    def __init__ (self, factor):
        self.factor = factor
    def __call__ (self,number):
        return number * self.factor
double = Multiplier (3)
print (double(10))
```




******### __iter__()  & __next__() method**

The __iter__ method enables an object to be iterable, allowing it to be used directly in for loops, list comprehensions, and functions like map(). When invoked, this method must return an iterator object (often self if the class implements the iteration protocol, or a separate iterator) that defines how to traverse the object's collection of data.

The __next__ method works hand-in-hand with __iter__ to define the exact mechanism for returning the next item in the sequence each time the loop advances. To prevent a loop from running indefinitely, this method is responsible for raising a StopIteration exception once all elements in the collection have been successfully traversed.

```python
#iteration support __iter__() & __next__()
class Counter:
    def __init__ (self, limit) :
        self.limit = limit
        self.current = 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value= self.current
        self.current += 1
        return value

for num in Counter(6):
    print(num)
```





### **__bool__ value method**

The __bool__ method defines the truth value of an object, dictating how it behaves when evaluated directly in a conditional statement like an if expression or when passed to the built-in bool() function. By default, custom Python objects are considered "truthy" unless this method is explicitly defined to return False based on specific internal conditions, such as an empty internal state or a zero value.

```python
# boolean value __bool__()
class BankAccount:
    def __init__ (self, balance):
        self.balance = balance
    def __bool__(self):
        return self.balance > 0
account = BankAccount(450)
if account:
    print("Account is active")
```

### **__del__() method**

It serves as an object destructor in python, which is automtically invoked by the garbage collector when an object's reference count drops to zero and is about to be completely removed from memory.

```python
#destructor method __del__()
class FileHandler:
    def __init__ (self, filename):
        self.filename = filename
        print("File opened")

    def __del__(self):
        print("File closed")
f = FileHandler ("data.txt")
```
